```python

for file, target in zip(file_list,target_list):
    LEV_list = []
    
    print(target)
    #print(file)
    df = pd.read_excel(path_dir+file)
    com_list = df["종목명"].values
    for com in com_list:
        #print(com)
        #print("len:",len(finance_file[(finance_file["종목명"]==com)]))
        if len(finance_file[(finance_file["종목명"]==com)]) == 0:
            LEV_list.append(None)
            continue
        finance = finance_file[(finance_file["종목명"]==com) & (finance_file["계정과목"]=="자산")][target]
        long_dis =all_df[(all_df["종목명"]==com)& (all_df["계정과목"]=="    장기차입금")][target]
        if finance.isna().values or long_dis.isna().values or len(finance)==0 or len(long_dis)==0:
            LEV_list.append(None)
            continue
#         if(com == "AK홀딩스"):
#             print(long_dis.values,finance.values)
#             print(target, file)
#             break
        LEV = long_dis.values/finance.values
        LEV_list.extend(LEV)
        #print(LEV)
    df["LEV"] = LEV_list    
    LIQ_list = []
    for com in com_list:
        #print(com)
        if len(all_df[all_df["종목명"]==com]) == 0:
            LIQ_list.append(None)
            continue
        cur_asset = all_df[(all_df["종목명"]==com)& (all_df["계정과목"]=="  유동자산")][target]
       # print(cur_asset)
        cur_debt = all_df[(all_df["종목명"]==com) & (all_df["계정과목"]=="  유동부채")][target]
        if cur_asset.isna().values or cur_debt.isna().values or len(cur_asset)==0 or len(cur_debt)==0:
            LIQ_list.append(None)
            continue
        LIQ = cur_asset.values/cur_debt.values
        LIQ_list.extend(LIQ)
    df["LIQ"] = LIQ_list
            
        
    df.to_excel("new_"+file)
    #print(LEV_list)
    
df
#len(LEV_list)
```
